```python
import pandas as pd
```

```python
import random
lst = ['robot'] * 10
lst += ['human'] * 10
lst += ['kiborg'] * 10
lst += ['machine'] * 10
random.shuffle(lst)
data = pd.DataFrame(lst, columns={'whoAmI': lst})
data.head(20)
```

```python
data = pd.DataFrame(random.sample(['human', 'robot']*10, 20) ,columns={'whoAmI': lst})
data.head(15)
```

```python
data.loc[data['whoAmI']!='human','human'] = '0'
data.loc[data['whoAmI']=='human','human'] = '1'
data.loc[data['whoAmI']!='robot','robot'] = '0'
data.loc[data['whoAmI']=='robot','robot'] ='1'
data.head(15)
```
